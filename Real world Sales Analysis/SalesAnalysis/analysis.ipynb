# Oues2: What city has the highest number of sell

```python
#Step1:add a city column
```

```python
all_data
```

```python
#Extracting city value from the address
```

```python
def find_city(address):
    return address.split(',')[1]


def find_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['city']=all_data["Purchase Address"].apply(lambda x:find_city(x)+ ' ' + find_state(x))    ####Here x represent all the row in permanent address
```

```python
all_data
```

```python
result=all_data.groupby('city').sum()
```

```python
result
```

```python
#Plotting the result
```

```python
import matplotlib.pyplot as plt

Cities=[city for city,df in all_data.groupby('city')]
plt.bar(Cities,result['sales'])
plt.xticks(Cities,rotation='vertical',size=10)
plt.xlabel('City')
plt.ylabel('Total sales')
plt.show()
```

# Oues4: Which product are often sold together?

```python
all_data.head()
```

```python
temp=all_data[all_data["Order ID"].duplicated(keep=False)]
```

```python
temp.head(20)
```

```python
temp['grouped']=temp.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
temp=temp[['Order ID','grouped']].drop_duplicates()
temp.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count=Counter()
for row in temp['grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))

    
for key,value in count.most_common(10):
    print(key,value)
```
